```python
def linear_regression(df):
    X,y=df[['X']],df['Y']
    model=LinearRegression()
    model.fit(X,y)
    y_pred=model.predict(X)
    plt.scatter(X,y,label='Data')
    plt.plot(X,y_pred,color='red',label='Linear Regression')
    plt.legend()
    plt.title("Linear Regression")
    plt.show()
linear_regression(df_linear)
```
